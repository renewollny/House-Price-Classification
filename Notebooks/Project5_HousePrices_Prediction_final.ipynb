```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline

df = pd.read_csv(r"/Volumes/Extreme Pro/WBS-Project 5/housing-classification-iter6.csv")
df.head(10)
```

```python
df.info()
```

```python
y = df.pop("Expensive")
y
```

```python
y.info()
```

```python
X = df
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
y_train.info()
```

```python
y_test.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

# select categorical and numerical column names
X_cat_columns = X.select_dtypes(exclude="number").copy().columns
X_num_columns = X.select_dtypes(include="number").copy().columns

# create numerical pipeline, only with the SimpleImputer(strategy="mean")
numeric_pipe = make_pipeline(
    SimpleImputer(strategy="mean"))
 
 # create categorical pipeline, with the SimpleImputer(fill_value="most_frequent") and the OneHotEncoder
categoric_pipe = make_pipeline(
    SimpleImputer(strategy="constant", fill_value="most_frequent"),
    OneHotEncoder(drop="first", handle_unknown="ignore")
)
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ("num_pipe", numeric_pipe, X_num_columns),
        ("cat_pipe", categoric_pipe, X_cat_columns),
    ]
)
```

```python
from sklearn.model_selection import GridSearchCV

full_pipeline = make_pipeline(preprocessor, 
                              DecisionTreeClassifier())

param_grid = {                                                                      # put in any parameters, there is
    "columntransformer__num_pipe__simpleimputer__strategy":["mean", "median"],
    "decisiontreeclassifier__max_depth": range(0, 14, 1),
    "decisiontreeclassifier__min_samples_leaf": range(0, 12, 1)
}

search = GridSearchCV(full_pipeline,
                      param_grid,
                      cv=5,
                      verbose=1)

search.fit(X_train, y_train)
```

```python
search.best_params_
```

```python
search.best_score_
```

```python
search.predict(X_test)
```

```python
y_pred_tree_test = search.predict(X_test)

from sklearn.metrics import accuracy_score

test_tree_accuracy = accuracy_score(y_true = y_test,
                    y_pred = y_pred_tree_test
                    )

round(test_tree_accuracy, 2)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_tree_test))
```
